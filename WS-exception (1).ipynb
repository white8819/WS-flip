### Q5

```python
driver = webdriver.Chrome('C:/Users/DC/Desktop/chromedriver.exe')
driver.maximize_window()

driver.get('https://github.com')
time.sleep(10)

driver.get(driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a').get_attribute('href'))
time.sleep(10)

table = driver.find_elements_by_xpath('//*[@class="Box-row"]')
title = []
desc = []
lng = []
stars = []

for row in table:
    title.append(row.find_element_by_xpath('./h1').text)
    try:
        desc.append(row.find_element_by_xpath('./p').text)
    except NoSuchElementException:
        desc.append('na')
        
    try:
        lng.append(row.find_element_by_xpath('./div[2]/span[1]/span[2]').text)
    except NoSuchElementException:
        lng.append('na')
        
    stars.append(row.find_element_by_xpath('./div[2]/a[2]').text)

print(len(title), len(desc), len(lng))
df = pd.DataFrame()
df['repo_title'] = title
df['repo_desc'] = desc
df['contri_cnt'] = stars
df['lng'] = lng

driver.close()
df.head()
```
