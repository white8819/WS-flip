```python
#importing libraries
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook as tqdm
import time

#Using proxy chrome driver
driver = webdriver.Chrome('C:/Users/DC/Desktop/chromedriver.exe')
```

Now, a function is created to scrape data from 5 pages

```python
def scrape_monster(address):
    
    driver.get(address) 
    time.sleep(5)
    content=driver.page_source
    soup=BeautifulSoup(content)
    
    role=[]
    company=[]
    location=[]
    experience=[]
    salary=[]
    jd=[]
    skills=[]

    cards=soup.findAll('div',attrs={'class':'card-apply-content'})
    for card in cards:

        # roles
        try:
            role.append(card.findAll('h3',attrs={'class':'medium'})[0].text)
        except:
            role.append('NA')

        # company name
        try:
            company.append(card.findAll('span',attrs={'class':'company-name'})[0].text)
        except:
            company.append('NA')

        # loc
        try:
            location.append(card.findAll('div',attrs={'class':'col-xxs-12 col-sm-5 text-ellipsis'})[0].text)
        except:
            location.append('NA')

        # exp
        try:
            experience.append(card.findAll('div',attrs={'class':'exp col-xxs-12 col-sm-3 text-ellipsis'})[0].text)
        except:
            experience.append('NA')

        # salary
        try:
            salary.append(card.findAll('div',attrs={'class':'package col-xxs-12 col-sm-4 text-ellipsis'})[0].text)
        except:
            salary.append('NA')

        # jd
        try:
            jd.append(card.findAll('p',attrs={'class':'job-descrip'})[0].text)
        except:
            jd.append('NA')

        # skills
        try:
            skills.append(card.findAll('p',attrs={'class':'descrip-skills'})[0].text)
        except:
            skills.append('NA')

    for i in range(0,len(skills)):
        skills[i]=skills[i].replace(' ','')
        skills[i]=skills[i].replace('\n','')
        skills[i]=skills[i].replace('skills:','')
    for i in range(0,len(experience)):
        experience[i]=experience[i].replace(' ','')
        experience[i]=experience[i].replace('\n','')
    for i in range(0,len(location)):
        location[i]=location[i].replace(' ','')
        location[i]=location[i].replace('\n','')
    for i in range(0,len(salary)):
        salary[i]=salary[i].replace(' ','')
        salary[i]=salary[i].replace('\n','')


    df = pd.DataFrame({'role':role,'comapany':company,'location':location,'experience':experience,'salary':salary,'JD':jd,'skills':skills})
    return df
```

```python
final_df = pd.DataFrame()
for i in tqdm(range(0,101,25)):
    df = scrape_monster(f'https://www.monsterindia.com/srp/results?start={i}&sort=1&limit=25&query=Software%20Engineer')
    final_df=pd.concat([final_df,df]).reset_index(drop=True)
    print(final_df.shape)
```
