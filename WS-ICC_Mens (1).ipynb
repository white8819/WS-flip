```python
#importing libraries
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
#Using proxy chrome driver
driver = webdriver.Chrome('C:/Users/DC/Desktop/chromedriver.exe')
```

```python
#reading data and parsing

def scrape_team(address):
    driver.get(address) 
    content=driver.page_source
    soup=BeautifulSoup(content)

    rows = soup.find_all('tr')

    df = []
    for row in rows[1:11]:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols[1:]]
        df.append(cols)

    df1 = pd.DataFrame(df, columns=['Country', 'Matches', 'Points', 'Rating'])
    df1['Country'] = df1.Country.str.split('\n').str[0]
    return df1
```

```python
df1=scrape_team('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
df1
```

```python
#Function2

def scrape_bat(address):
    driver.get(address) 
    content=driver.page_source
    soup=BeautifulSoup(content)


    rows = soup.find_all('tr')

    df = []
    for row in rows[1:11]:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols[1:4]]
        df.append(cols)

    df2 = pd.DataFrame(df, columns=['Player', 'Team', 'Rating'])
    return df2
```

```python
df2=scrape_bat('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
df2
```
